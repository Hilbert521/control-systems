# Hamilton Jacobi Bellman Equation

Modern optimal control theory is heavily reliant on the Hamilton Jacobi Bellman equation. This equation can be derived from the Bellman's prinicple of optimality:

>An optimal policy has the property that no matter what choice is made initially, the resulting choices from the state after the first choise must be optimal.

Let $F(x,a)$ be the cost of taking action $a$ at state $x$. Then the total cost can be expressed as:

$$V(x_0) = \min_{\{a_k\}_{k=0}^N} \sum_{k=0}^N F(x_k, a_k)$$

Now we can separate the first step from the remaining steps. This is equivalent to Bellman's principle of optimality.

$$ V(x_0) = \min_{a_0} \left( F(x_0, a_0) + \min_{\{a_k\}_{k=1}^N}  \sum_{k=1}^N  F(x_k, a_k) \right)$$

We now note that this is a recursive equation.

$$ V(x_0) = \min_{a_0} \left( F(x_0, a_0) + V(x_1) \right)$$

This equation can be solved using dynamic programming.

1. Start at the desired end point.
1. Perform all possible actions at the end point, recording their cost.
1. For the resulting states of the performed actions, again perform all actions.
1. If a state can be reached by more than one sequence of actions, choose the action sequence with the lower cost.
1. Repeat the process until the initial state is reached by an action sequence whose cost is lower than an other action sequences that may reach the initial state.

Let us solve the problem of computing the shortest path around a hilly terrain. We will represent the terrain using the addition of several functions in 2D.

```python
n = 20
x = linspace(-1,1,n)
y = linspace(-1,1,n)
nx = len(x)
ny = len(y)
dx = 2.0/n
dy = 2.0/n

def terrain(x):
    return cos(x[0]*pi/2)*cos(x[1]*pi/2) - 0.2*exp(1.0/(0.5 + (x[0]-0.5)**2 + (x[1]-0.5)**2))

z = zeros([nx,ny])
for i in range(nx):
    for j in range(ny):
        z[i,j] = terrain([x[i], y[j]])
X,Y = meshgrid(x, y)
ax = subplot(111, projection='3d')
h_terrain = ax.plot_wireframe(X, Y, z, cstride=1, rstride=1)

def position(i_x):
    return array([X[i_x[0], i_x[1]], Y[i_x[0],i_x[1]]])

i_x0 = array([0,0])
i_xf = array([n-1,n-1])
x0 = position(i_x0)
xf = position(i_xf)

h_start = ax.plot3D([x0[0]], [x0[1]], [terrain(x0)], 'go')
h_end = ax.plot3D([xf[0]], [xf[1]], [terrain(xf)], 'ro')
ax.view_init(elev=60, azim=-50)
ax.legend( ['start','end','terrain'])
```
